```python
%run ./Corpus_Reader.ipynb
%run ./POS_Tagger.ipynb
%run ./Grammar.ipynb
import tkinter
from tkinter import *
from PIL import ImageTk, Image
import string

def IsAmharCharacterorNot(text):
    
    englsihChar=string.ascii_letters
    
    return all((False if char in englsihChar else True for char in text))
def SentenceParser():
    
    #check if the text entry is empty or not
    if  txt.get():
        
        lblError.config(text="")  
        
        #check if the text enty is amharic or not 
        if IsAmharCharacterorNot(txt.get())==False:
            
             ErrorMessage("እባክዎን የአማርኛ ቃላት ብቻ ያስገቡ")
                
        else: 

            regexp_tagger = nltk.RegexpTagger(POS_Tagger)

            sentence=regexp_tagger.tag(nltk.word_tokenize(txt.get()))
            
            #this is to know number of words

            wordtokenize=nltk.word_tokenize(txt.get()+"\n")


            verbNumber=0
            

            #check if there is verb in the sentence or not
            for  i in range(0,len(sentence),1): 

                j=0
               
                for data in sentence[i]:

                    if j%2!=0:

                        if data=="V":

                            verbNumber=1

                    j=j+1

            if(verbNumber==0):

                ErrorMessage("ያስገቡ ዓረፍተ ነገር ግስ(verb) አይታወቅም ውይም የለውም")
                
            #check if the number of word is two
            elif(len(wordtokenize)==2):

                Reg_parser=nltk.RegexpParser(twoWordGrammar) 

                parseDraw(sentence,Reg_parser)
                
                print("two Word Grammar")
                
             #check if the number of word is three
            elif(len(wordtokenize)==3):

                Reg_parser=nltk.RegexpParser(threeWordGrammar) 

                parseDraw(sentence,Reg_parser)
                
                print("three Word Grammar")
            
            #check if the number of word is four
            elif(len(wordtokenize)==4):

                Reg_parser=nltk.RegexpParser(fourWordGrammar) 

                parseDraw(sentence,Reg_parser)
                
                print("four Word Grammar")

             #check if the number of word is five
            elif(len(wordtokenize)==5):

                Reg_parser=nltk.RegexpParser(fiveWordGrammar) 

                parseDraw(sentence,Reg_parser)
                
                print("five Word Grammar")
                
              #check if the number of word is six
            elif(len(wordtokenize)==6):
                
                Reg_parser=nltk.RegexpParser(sixWordGrammar) 

                parseDraw(sentence,Reg_parser)
                
            elif(len(wordtokenize)==7):
                
                Reg_parser=nltk.RegexpParser(sevenWordGrammar) 

                parseDraw(sentence,Reg_parser)
                

            else:

                Reg_parser=nltk.RegexpParser(GeneralGrammar) 

                parseDraw(sentence,Reg_parser)

    else:
        ErrorMessage("እባክዎን ዓረፍተ ነገርዎትን ያስገቡ")
        
def parseDraw(sentence,Reg_parser):
    
    Reg_parser.parse(sentence) 

    Output=Reg_parser.parse(sentence)
    
    temp=0;
    
     #check if the parser divide a sentence using NP and VP like a grammar 
    for data in Output:
        
        temp=temp+1      
        
    if(temp==2):
         
        write_text = open("Tested_Log.txt","a",encoding="utf8")

        write_text.write(txt.get())

        write_text.write("\n")

        write_text.close()
        
        Output.draw()
        
    else:
        
        ErrorMessage("እባክዎን ዓረፍተ ነገርዎትን ያስተካክሉ")        
        
        write_text = open("WrongSentence_Log.txt","a",encoding="utf8")

        write_text.write(txt.get())

        write_text.write("\n")

        write_text.close()
        
        Output.draw()
        
def ErrorMessage(message):
    
    lblError.config(text=message)
                   
window = Tk()

window.title("Welcome to Parsing Tree")

window.geometry('800x500')

lbl1 = Label(window, text="የ አማርኛ ዓረፍተ ነገር ፓርሰር",font=(50),fg="green")

lbl1.grid(column=1, row=1,padx=10, pady=10)

lbl = Label(window, text="ዓረፍተ ነገር ያስገቡ",font=(18),fg="green")

lbl.grid(column=0, row=2,padx=20, pady=20)

lbl.grid(column=0, row=2,padx=20, pady=20)

txt = Entry(window,width=70)

txt.grid(column=1, row=2,pady=50)

btn = Button(window, text="ፓርስ አድርግ",font=(13), command=SentenceParser, bg="green", fg="white")

btn.grid(column=2, row=2,padx=20)

lblError=Label(window, text=" ",font=(18), fg="red")

lblError.grid(column=1, row=3,padx=0, pady=0)

path = "Capture.JPG"

img = ImageTk.PhotoImage(Image.open(path))

imageLable =tkinter.Label(window, image = img)

imageLable.grid(column=1, row=4,padx=0, pady=0)

window.mainloop()
```
